```python
from pandas.plotting import autocorrelation_plot
```

```python
#The autocorrelation_plot shows that correlation decreases to 0 as lag increases
```

```python
from statsmodels.graphics.tsaplots import plot_acf

store1_open = store1[store1.Open == 1]

plot_acf(store1_open['Sales'], lags=10)
```

```python
plot_acf(store1_open['Sales'], lags=25)
```

```python
from statsmodels.tsa.arima_model import ARMA

store1_sales = store1_open[['Sales']].astype(float)

model = ARMA(store1_sales, (1, 0)).fit()
model.summary()
```

```python
model = ARMA(store1_sales, (2, 0)).fit()
model.summary()
```

```python
model.resid.plot()
```

```python
#The residuals show that our model does not account well during the holiday spikes in sales
```

```python
#expand this AR model to an ARMA model, including the moving average

model = ARMA(store1_sales, (1, 1)).fit()
model.summary()
```

```python
# The AR coefficient represents dependency on the last value 
# and the MA component represents any spikes independent of the last value.
```

```python
# We can also fit a true ARIMA model to predict the difference of the series.

from statsmodels.tsa.arima_model import ARIMA

model = ARMA(store1_sales, (1, 1, 1)).fit()
model.summary()
```

```python
# We can remove the MA component since it does not appear to be useful.
# This is due to a low coef and the percentile range includes 0
model = ARIMA(store1_sales, (1, 1, 0)).fit()
model.summary()
```

```python
# We can compare the last 50 days of true values against our predictions.
model.plot_predict(1, 50)
```
