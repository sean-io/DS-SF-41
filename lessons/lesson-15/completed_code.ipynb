```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot
```

```python
df = pd.read_csv('../../data/rossmann.csv', skipinitialspace=True, low_memory=False)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
```

```python
df.head()
```

```python
store1 = df[df['Store'] == 1]
```

```python
store1[store1['Open'] == 1].plot(y='Sales', figsize=(10,8))
```

```python
df['Sales'].resample('D').mean().autocorr(lag=2)
```

```python
store1[store1['Open'] == 1]['Sales'].resample('W').mean().plot()
```

```python
store1['Sales'].rolling(window=30, center=True).mean().plot()
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
autocorrelation_plot(store1['Sales'])
```

```python
from statsmodels.graphics.tsaplots import plot_acf
store1_open = store1[store1['Open'] == 1]
plot_acf(store1_open['Sales'], lags=10)
```

```python
plot_acf(store1_open['Sales'], lags=25)
```

```python
from statsmodels.tsa.arima_model import ARMA

store1_sales = store1_open[['Sales']].astype(float)
model = ARMA(store1_sales, (1, 0)).fit()
model.summary()
```

```python
model = ARMA(store1_sales, (2, 0)).fit()
model.summary()
```

```python
model.resid.plot()
```

```python
plot_acf(model.resid, lags=50)
```

```python
model = ARMA(store1_sales, (1, 1)).fit()
model.summary()
```

```python
model = ARMA(store1_sales, (1, 0, 1)).fit()
model.summary()
```

```python
model.resid.plot()
```

```python
plot_acf(model.resid, lags=50)
```

```python
model = ARMA(store1_sales, (1, 1, 1)).fit()
model.summary()
```

```python
model.plot_predict(0, 50)
```

```python
model = ARMA(store1_sales, (7, 1, 2)).fit()
model.summary()
```

```python
model.resid.plot()
```

```python
plot_acf(model.resid, lags=50)
```

```python
model.plot_predict(0, 200)
```
